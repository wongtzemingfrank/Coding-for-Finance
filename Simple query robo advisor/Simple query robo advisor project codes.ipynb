```python
portfolios_fn = "portfolios.csv"
price_fn = "price_data.csv"
target_stock = "NYSEARCA:JKH"
purchase_date = "2016-09-30"
```

```python
import pandas as pd
import math
```

```python
portfolio_df = pd.read_csv(portfolios_fn)
portfolio_df.head()
```

```python
price_df = pd.read_csv(price_fn)
price_df = price_df.set_index('date')
price_df.head()
```

```python
price_of_target_stock = price_df.loc[purchase_date, target_stock] #price of target_stock at purchase_date
```

```python
column_matcher = {"x_1" : "w_1", "x_2" : "w_2", "x_3" : "w_3", "x_4" : "w_4", "x_5" : "w_5", "x_6" : "w_6", "x_7" : "w_7"}
```

```python
#Idea is to go through the portfolio_df using for loop.
#Evaluate if there is target_stock within
#If there is, figure out the column name, which can be x_1 ... x_7
#Figure out the weightage for example w_7
#Figure out the invt_amt
#that particular portfolio will have target_stock_count = math.trunc((invt_amt * weightage) / price_of_target_stock)
#add target_stock_count to total_num_of_target_stocks
```

```python
total_num_of_target_stocks = 0
for i in range(portfolio_df.shape[0]):
    if target_stock in list(portfolio_df.iloc[i, :]):
        invt_amt = list(portfolio_df.iloc[i, :])[1]
        index_of_target_stock = list(portfolio_df.iloc[i, :]).index(target_stock)
        weightage_of_target_stock = list(portfolio_df.iloc[i, :])[index_of_target_stock + 7]
        total_num_of_target_stocks += math.trunc((invt_amt * weightage_of_target_stock) / price_of_target_stock)
```
