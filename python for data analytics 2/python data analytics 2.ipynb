## Data processing, 4 steps:

first step

```python
#Now I need to convert rating, gender, education, marital, S(n) into category data type:
data['RATING'] = data['RATING'].astype('category')
data['GENDER'] = data['GENDER'].astype('category')
data['EDUCATION'] = data['EDUCATION'].astype('category')
data['MARITAL'] = data['MARITAL'].astype('category')
data['S1'] = data['S1'].astype('category')
data['S2'] = data['S2'].astype('category')
data['S3'] = data['S3'].astype('category')
data['S4'] = data['S4'].astype('category')
data['S5'] = data['S5'].astype('category')
```

second step

```python
#Knowing that R3 is object dtype, it means there is a mixture of data
data['R3'] = data['R3'].str.replace('$', '')
data['R3'] = data['R3'].str.replace(',', '')
```

third step

```python
#Convert R3 into integer type
data['R3'] = data['S1'].astype('int')
```

```python
data.info()
```

forth step

```python
#Knowing that there are null values in pandas, and since these null values are quite little, it can be dropped with affecting
#the credibility of my dataset study
data = data.dropna()
```
